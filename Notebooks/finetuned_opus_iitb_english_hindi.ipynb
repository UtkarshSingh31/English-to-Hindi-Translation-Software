## Local Inference on GPU
Model page: https://huggingface.co/shogun-the-great/finetuned-opus-iitb-english-hindi

⚠️ If the generated code snippets do not work, please open an issue on either the [model repo](https://huggingface.co/shogun-the-great/finetuned-opus-iitb-english-hindi)
			and/or on [huggingface.js](https://github.com/huggingface/huggingface.js/blob/main/packages/tasks/src/model-libraries-snippets.ts) 🙏

```python
from transformers import MarianMTModel, MarianTokenizer

# Load the model and tokenizer
model_name = "Helsinki-NLP/opus-mt-en-hi"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def translate(text):
    # Tokenize
    encoded = tokenizer(text, return_tensors="pt", padding=True)

    # Generate translation
    translated = model.generate(**encoded, num_beams=5)

    # Decode to text
    return tokenizer.decode(translated[0], skip_special_tokens=True)

```

```python
# Test sentences
sentences = [
    "Do not enter this area.",
    "Keep the environment clean.",
    "Parking is not allowed here.",
    "This is a public notice.",
    "Smoking is strictly prohibited.",
    "This is to inform the public that the office will remain closed on 12th December for maintenance.",
    "A brown wallet was found near Gate 3 on 5th December. The rightful owner may claim it with valid identification.",
    ""

]

for s in sentences:
    print(f"EN: {s}")
    print(f"HI: {translate(s)}")
    print()
```
